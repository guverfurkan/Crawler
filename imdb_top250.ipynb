```python
#installation of required libraries
import requests
from bs4 import BeautifulSoup
import pandas as pd
import re,json
```

```python
#listing required links
url_list=['https://www.imdb.com/search/title/?groups=top_250&sort=user_rating,desc']

for i in ('51','101','151','201'):
    url_list.append(f'https://www.imdb.com/search/title/?groups=top_250&sort=user_rating,desc&start={i}&ref_=adv_nxt')
```

```python
#function for pulling, categorizing and saving data
no=[]
name=[]
year=[]

def crawl(url):
    r=requests.get(url)
    soup=BeautifulSoup(r.content,'html.parser')
    gelen_veri=soup.find_all("h3",{"class":"lister-item-header"})
    soup2=BeautifulSoup(str(gelen_veri),"html.parser")

    
    for i in soup2.find_all("span",{ "class" : "lister-item-index unbold text-primary"}):
        no.append(i.text.replace('.',''))
    
    
    for i in soup2.find_all("a"):
           name.append(i.text)

    
    for i in soup2.find_all("span",{ "class" : "lister-item-year text-muted unbold"}):
           year.append(i.text.replace('(','').replace(')',''))
    
    return(no,name,year)
```

```python
#apply function for all url's
for url in url_list:
    crawl(url)
```

```python
imdb=pd.DataFrame({'no':no,'name':name,'year':year})
imdb
```
